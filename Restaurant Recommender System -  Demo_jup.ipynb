```python
''' 
Taking input from the user
'''
inp_rest_id = int(input("Enter the restaurant ID: "))
online_available = int(input("\nShould the restaurant have Onlne order facilities: \n1.Yes \n2.No\nEnter your response: "))

budget = int(input("\nFilter according to budget?: \n1.Yes \n2.Skip\nEnter your response: "))
if budget == 1:
    upper_limit = float(input("Enter the budget. Upper limit: "))
else:
    upper_limit = 10000
    
inp_rest = rest_data_transformed[(rest_data_transformed['rest_id'] == inp_rest_id)]
print("\nDetails of input restaurant: \n", inp_rest[['name', 'rest_id', 'rate', 'rating_encoded', 'location']])

recommended_rest = pd.DataFrame()

rec = recommend_class.reccommend(inp_rest_id, upper_limit, model_knn, rest_data_transformed, recommended_rest)

if (online_available == 1) & (budget == 1):
      recommended_rest = rec.recommend_according_to_budget_and_online(inp_rest_id, upper_limit, model_knn, rest_data_transformed, recommended_rest)
elif (online_available != 1) & (budget == 1):
      recommended_rest = rec.recommend_according_to_budget(inp_rest_id, upper_limit, model_knn, rest_data_transformed, recommended_rest)
elif (online_available == 1) & (budget != 1):
      recommended_rest = rec.recommend_online(inp_rest_id, model_knn, rest_data_transformed, recommended_rest)
elif (online_available != 1) & (budget != 1):
      recommended_rest = rec.recommend_rating(inp_rest_id, model_knn, rest_data_transformed, recommended_rest)

print("\nSeven Restaurant Recommendations Similar to given retaarant: \n")
recommended_rest[['name', 'rest_id', 'rate', 'rating_encoded', 'location', 'rest_type', 'cuisines', 'approx_cost(for two people)']]
```
